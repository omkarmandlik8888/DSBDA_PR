```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
words = set(nltk.corpus.words.words())
```

```python
sentence = df2['tweet'][0]
sid.polarity_scores(sentence)['compound']
```

```python
import re
def cleaner(tweet):
    tweet = re.sub("@[A-Za-z0-9]+","",tweet) #Remove @ sign
    tweet = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", tweet) #Remove http links
    tweet = " ".join(tweet.split())
    tweet = tweet.replace("#", "").replace("_", " ") #Remove hashtag sign but keep the text
    tweet = " ".join(w for w in nltk.wordpunct_tokenize(tweet)
         if w.lower() in words or not w.isalpha())
    return tweet
    

df2['tweet_clean'] = df2['tweet'].apply(cleaner)
```

```python
list1 = []
for i in df2['tweet_clean']:
    list1.append((sid.polarity_scores(str(i)))['compound'])
```

```python
df2['sentiment'] = pd.Series(list1)

def sentiment_category(sentiment):
    label = ''
    if(sentiment>0):
        label = 'positive'
    elif(sentiment == 0):
        label = 'neutral'
    else:
        label = 'negative'
    return(label)

df2['sentiment_category'] = df2['sentiment'].apply(sentiment_category)
```

```python
df2.head()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
positive = df2[df2['sentiment_category']=='positive']
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(str(positive['tweet']))
plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
negative = df2[df2['sentiment_category']=='negative']
wordcloud = WordCloud(max_font_size=50, max_words=500, background_color="white").generate(str(negative['tweet']))
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.show()
```
